# Code to calculate distance from one country (its capital) to another

```python
import csv
from geopy.distance import great_circle
import pandas as pd
```

```python
capitals = pd.read_csv("country_to_continent.csv")

def calculate_distances():
    distance_data = []

    for index_from, capital_from in capitals.iterrows():
        from_coords = (capital_from["Latitude"], capital_from["Longitude"])

        for index_to, capital_to in capitals.iterrows():
            if capital_from["Capital"] != capital_to["Capital"]:
                to_coords = (capital_to["Latitude"], capital_to["Longitude"])
                distance = great_circle(from_coords, to_coords).kilometers
                distance_data.append({
                    "country_from": capital_from["Country"],
                    "capital_from": capital_from["Capital"],
                    "country_to": capital_to["Country"],
                    "capital_to": capital_to["Capital"],
                    "distance_in_km": distance
                })

    return pd.DataFrame(distance_data)
```

```python
# Generate the dataframe with all combinations of distances
distance_df = calculate_distances()

# Save the dataframe to a new CSV file if needed
distance_df.to_csv('country_distances.csv', index=False)
```
