# Determine if a country is an island or not

To use it to consider ways of trading

```python
import pandas as pd
import numpy as np
```

```python
distance_df = pd.read_csv(r"..\data\FAOSTAT\clean_data\food_trading\country_distances_matrix.csv")
```

```python
distance_df.head()
```

```python
border_nr_df = distance_df.groupby('country_from')['share_border'].sum().to_frame() #number of border countries (including itself)
```

```python
border_nr_df.sort_values(by='share_border', ascending=True)
```

If only shares border with itself = Island or missing info

```python
border_nr_df[border_nr_df['share_border'] == 1]
```

```python
no_border_countries = border_nr_df[border_nr_df['share_border'] == 1].index.to_list()

island_countries_chatGPT = [ #country islands from chatGPT
    "Antigua and Barbuda",
    "Bahamas",
    "Bahrain",
    "Barbados",
    "Cape Verde (Cabo Verde)",
    "Comoros",
    "Cuba",
    "Cyprus",
    "Dominica",
    "Dominican Republic",
    "East Timor (Timor-Leste)",
    "Fiji",
    "Grenada",
    "Iceland",
    "Indonesia",
    "Ireland",
    "Jamaica",
    "Japan",
    "Kiribati",
    "Madagascar",
    "Maldives",
    "Malta",
    "Marshall Islands",
    "Mauritius",
    "Micronesia (Federated States of Micronesia)",
    "Nauru",
    "New Zealand",
    "Palau",
    "Papua New Guinea",
    "Philippines",
    "Samoa",
    "São Tomé and Príncipe",
    "Seychelles",
    "Singapore",
    "Solomon Islands",
    "Sri Lanka",
    "St. Kitts and Nevis",
    "St. Lucia",
    "St. Vincent and the Grenadines",
    "Tonga",
    "Trinidad and Tobago",
    "Tuvalu",
    "United Kingdom",
    "Vanuatu"
]
```

```python
island_countries_chatGPT.extend(no_border_countries)
all_islands_ormissing_list = [item.lower() for item in island_countries_chatGPT]
```

```python
distance_df['any_island_or_missing'] = np.where(distance_df['country_from'].isin(set(all_islands_ormissing_list)) | distance_df['country_to'].isin(set(all_islands_ormissing_list)), 1, 0)
```

```python
distance_df.to_csv(r"..\data\FAOSTAT\clean_data\food_trading\country_distances_matrix_border_island.csv", index=False)
```
