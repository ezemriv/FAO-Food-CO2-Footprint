```python
import geopandas as gpd

# Load the world dataset from Natural Earth
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
world
```

```python

# Filter out only the countries that are islands
islands = world[world['geometry'].is_empty]
```

```python
islands
```

```python

# Print the names of the island countries
print(islands['name'])
```
