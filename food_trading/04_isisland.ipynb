# Determine if a country is an island or not

To use it to consider ways of trading

```python
import pandas as pd
import numpy as np
```

```python
distance_df = pd.read_csv(r"..\data\FAOSTAT\clean_data\food_trading\country_distances_matrix.csv")
```

```python
distance_df.head()
```

```python
distance_df[distance_df['country_from'] == 'cuba']
```

```python
border_nr_df = distance_df.groupby('country_from')['share_border'].sum().to_frame() #number of border countries (including itself)
```

```python
border_nr_df.sort_values(by='share_border', ascending=True)
```

If only shares border with itself = Island or missing info

```python
pd.set_option('display.max_rows', None)

border_nr_df[border_nr_df['share_border'] == 1]
```

```python
islands_ormissing_list = border_nr_df[border_nr_df['share_border'] == 1].index.to_list()
```

```python
distance_df['island_or_missing'] = np.where(distance_df['country_from'].isin(islands_ormissing_list), 1, 0)
```

```python
distance_df.to_csv(r"..\data\FAOSTAT\clean_data\food_trading\country_distances_matrix.csv")
```
